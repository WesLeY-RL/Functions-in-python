```python
import numpy as np
```

```python
v = np.array([[1],[1]])
v
```

```python
# rotação em 45 graus
dgr = np.pi / 2
R = np.array([[np.cos(dgr), -np.sin(dgr)],
              [np.sin(dgr), np.cos(dgr)]])
R
```

```python
# Rotacionando
matrizv_rot = np.dot(R,v)
matrizv_rot
```

```python
#Vizualização

import matplotlib.pyplot as plt
origem  = np.array([0,0])
fig, ax = plt.subplots()
ax.quiver(*origem,*v, angles= "xy", scale_units= "xy", scale=1, color='r')
ax.quiver(*origem,*matrizv_rot, angles= "xy", scale_units= "xy", scale=1, color='g')
ax.set_xlim(-2, 3)
ax.set_ylim(-2, 3)
plt.grid(True)
plt.show()

```

```python


# Função para normalizar um vetor
def normalize(v):
    norm = np.linalg.norm(v)
    if norm == 0: 
       return v
    return v / norm

# Função para o procedimento de Gram-Schmidt
def gram_schmidt(vectors):
    orthonormal_basis = []
    for v in vectors:
        for b in orthonormal_basis:
            v = v - np.dot(v, b) * b
        orthonormal_basis.append(normalize(v))
    return orthonormal_basis

# Vetor inicial
v1 = np.array([3, 4, 5])
v1 = normalize(v1)

# Encontrar um vetor não colinear
v2 = np.array([1, 0, 0]) if v1[0] == 0 else np.array([0, 1, 0])

# Aplicar Gram-Schmidt
basis = gram_schmidt([v1, v2, np.cross(v1, v2)])

# Exibir a base ortonormal
for i, vec in enumerate(basis):
    print(f"Vetor {i+1}: {vec}")
```
